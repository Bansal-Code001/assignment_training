## Data visualisation -- graphs and plots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
arr1 = np.random.randint(2000,5000,100)
arr1
```

## line plot

```python
## line plot 
plt.plot(arr1)
```

```python
arr1.sort()
plt.plot(arr1)
```

```python
x = [1,2,3,4,5]
y = [7,20,4,9,17]
plt.plot(x,y,marker = "h")
```

```python
plt.plot(x,y,
         marker="o",
         ms="20",
         mfc="yellow",
         mec="black",
         color="violet",
         linewidth="4",
         ls="dashdot",
         alpha=0.5)
plt.xlabel("x axis")
plt.ylabel("y axis")
```

## Scatter plot

```python
# scatter plot 
x = [1,2,3,4,5]
y = [1,4,20,16,25]
plt.scatter(x,y,
            marker="2",
            )
```

```python
df = pd.read_csv("Used_Bikes.csv")
df.head(10)
```

```python
plt.scatter(df["kms_driven"],df["price"],marker="2")
```

## Bar plot

```python
# bar plot 
x = ["mp","up","rj","mh"]
y = [100,170,60,240]
colors =["red","green","yellow","blue"]
hatch_patterns = ["/","\\\\","//\\","o"]
plt.bar(x,y,color=colors,edgecolor="black",hatch=hatch_patterns,label=x)
plt.legend()
plt.grid()
```

## Histogram

```python
## histogram 
data = np.random.randn(1000)
plt.hist(data,bins=30)
plt.show()
```

## Pie chart

```python
## pie chart == distribution of data in slice 
labels = ["a","b","c","d","e","f","g"]
sizes = [10,2,8,5,18,35,22]
explodes = (0,0,0.4,0,0,0,0)
plt.pie(sizes,labels = labels, autopct= "%1.2f%%",shadow=True,explode=explodes)
plt.savefig("plot.png")
plt.show()
plt.title("pie chart")
```
